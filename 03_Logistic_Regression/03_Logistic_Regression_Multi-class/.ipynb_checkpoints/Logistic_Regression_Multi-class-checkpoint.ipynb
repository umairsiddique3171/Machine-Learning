```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target.shape
```

```python
digits.data.shape
```

```python
digits.feature_names
```

```python
digits.images[0] # 2d array of 8x8 pixels where values give pixel intensity
```

```python
digits.data[0]
```

```python
digits.images[0].flatten()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size = 0.25,random_state =10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.matshow(digits.images[42])
```

```python
digits.target[42]
```

```python
model.predict([digits.data[42]])
```

```python
y_predicted = model.predict(x_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
```

```python
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
data['target'] = iris.target
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data.drop('target',axis = 1),data.target,test_size = 0.1,random_state = 10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model1 = LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```
