```python
import numpy as np
```

```python
from scipy.stats import norm

plt.hist(df.height, bins = 20 ,rwidth = 0.8, density = True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(df.height.min(),df.height.max(),0.1)
plt.plot(rng,norm.pdf(rng,df.height.mean(),df.height.std()))
```

```python
df.height.mean()
```

```python
df.height.std()
```

```python
upper_limit = df.height.mean() + 3*df.height.std()
lower_limit = df.height.mean() - 3*df.height.std()
lower_limit,upper_limit
```

```python
df[(df.height>upper_limit) | (df.height<lower_limit)]
```

```python
new_data = df[(df.height<upper_limit) & (df.height>lower_limit)]
```

```python
df.shape[0] - new_data.shape[0]
```

Now using z-score to do the same thing

```python
df['zscore'] = (df.height - df.height.mean())/df.height.std()
```

```python
df[(df['zscore'] > 3) | (df['zscore'] < -3)]
```

first of all removing outliers using percentile technique. 

```python
min_threshold,max_threshold = df.price_per_sqft.quantile([0.001,0.999])
```

```python
min_threshold,max_threshold
```

```python
df[(df.price_per_sqft < min_threshold) | (df.price_per_sqft > max_threshold)]
```

```python
len(df[(df.price_per_sqft < min_threshold) | (df.price_per_sqft > max_threshold)])
```

```python
df1 = df[(df.price_per_sqft > min_threshold) & (df.price_per_sqft < max_threshold)]
```

```python
len(df1)
```

removing outliers from new dataframe using standard devitaion technique

```python
print(df1.price_per_sqft.mean())
print(df1.price_per_sqft.std())
```

```python
upperlimit = df1.price_per_sqft.mean() + 4*df1.price_per_sqft.std()
lowerlimit = df1.price_per_sqft.mean() - 4*df1.price_per_sqft.std()
lowerlimit,upperlimit
```

```python
df1[(df1.price_per_sqft < lowerlimit) | (df1.price_per_sqft > upperlimit)]
```

```python
len(df1[(df1.price_per_sqft < lowerlimit) | (df1.price_per_sqft > upperlimit)])
```

```python
df2 = df1[(df1.price_per_sqft > lowerlimit) & (df1.price_per_sqft < upperlimit)]
```

```python
df2.shape[0]
```

```python
df1.shape[0]-df2.shape[0]
```

now doing same using zscore technique

```python
df1['zscore'] = (df1.price_per_sqft - df1.price_per_sqft.mean())/df1.price_per_sqft.std()
```

```python
df1.head()
```

```python
df1[(df1.zscore < -4) | (df1.zscore > 4)]
```

```python
len(df1[(df1.zscore < -4) | (df1.zscore > 4)])
```

```python
df2 = df1[(df1.zscore > -4) & (df1.zscore < 4)]
```

```python
df2.shape[0]
```

```python
df1.shape[0]-df2.shape[0]
```

Now plotting histogram and bell shaped curve

```python
plt.hist(df2.price_per_sqft,bins = 20,rwidth = 0.7)
plt.xlabel('Price/sqft')
plt.ylabel('Count')
```

```python
from scipy.stats import norm

plt.hist(df2.price_per_sqft, bins = 20 ,rwidth = 0.8, density = True)
plt.xlabel('Price/sqft')
plt.ylabel('Count')

rng = np.arange(df2.price_per_sqft.min(),df2.price_per_sqft.max(),0.1)
plt.plot(rng,norm.pdf(rng,df2.price_per_sqft.mean(),df2.price_per_sqft.std()))
```
