```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x.head()
```

```python
y.head()
```

```python
# so rather than using onehotencoding --> labelencoding on y I am using y as target of iris data form sklearn
```

```python
y = iris.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
x_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(model,filled = True)
# wherever there is binary division in tree. We call it a cart tree.
```

```python
# now doing post-prunning on another model. Normally we do it on same model. But for clear understanding I am doing this.
```

```python
model1 = DecisionTreeClassifier(max_depth = 2)
# hyperparameter -> maxdepth = 2
```

```python
model1.fit(x_train,y_train)
```

```python
plt.figure(figsize = (15,10))
tree.plot_tree(model1,filled = True)
```

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
# checking accuracy using sklearn method
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model1.predict(x_test)
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score # notice it's same as given by model1.score()
```

```python
# checking classification report using sklearn method
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pred,y_test))
```
