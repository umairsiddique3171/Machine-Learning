```python
def meta_data(data):
    total = data.isnull().sum()
    percent = (total)/len(data)*100
    datatype = data.dtypes
    unique = data.nunique()
    
    return pd.concat([total, percent, datatype, unique],axis = 1 , keys = ['Total_missing_values','percent','datatypes','unique'])
```

```python
meta_data(data_lr)
```

```python
data_lr.drop(['CHAS','DIS','B'],axis = 1,inplace = True)
```

```python
data_lr.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = data_lr.drop('label',axis = 1)
```

```python
y = data_lr.label
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
linear.coef_
```

```python
linear.intercept_
```

```python
train_predict = linear.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_train = r2_score(y_train,train_predict)
```

```python
r2_train
```

```python
test_predict = linear.predict(x_test)
```

```python
r2_test = r2_score(y_test,test_predict)
```
